```python
%matplotlib notebook
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')
data=pd.read_csv("vico.csv")
#print(data)
vcount=data["view_count"]
like=data["likes"]
ratio=data["ratio"]
plt.scatter(vcount,like,c=ratio,cmap="summer",edgecolor="black",linewidth=1,alpha=0.75)
c_bar=plt.colorbar()
c_bar.set_label("Like/Dislike Ratio")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Viewcount")
plt.ylabel("Likes")
plt.tight_layout()
```
